```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import string 
```

```python
df = pd.read_csv('survey_results_public.csv')
pd.set_option('max_columns', None)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cols = ['EdLevel', 'Employment', 'JobSat', 'NEWDevOps', 'NEWDevOpsImpt', 
        'OpSys', 'SurveyEase', 'SurveyLength']      
```

```python
def plot_cols_with_converted_comp(df, columns):
    for column in columns:
        plt.figure(figsize = (15, df[column].nunique() / 2))
        df.groupby([column])['ConvertedComp'].mean().sort_values(). \
            tail(5).plot.barh()
```

```python
plot_cols_with_converted_comp(df, cols)
```

```python
def plot_cols_with_job_sat(df, columns):
    for column in columns:
        plt.figure()
        plt.xticks(rotation = 45, ha = 'right')
        sns.countplot(x = column, hue = 'JobSat', data = df, dodge = False) 
```

```python
plot_cols_with_job_sat(df, cols)
```
